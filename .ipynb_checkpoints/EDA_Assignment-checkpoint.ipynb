### Univariate Visualization
* Plot two Charts
* Provide each charts with at least two insights 

```python
sns.set(style="whitegrid")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(df["sepal.length"], kde=True, ax=axes[0])
axes[0].set_title("Histogram of Sepal Length")
axes[0].set_xlabel("Sepal Length")
axes[0].set_ylabel("Frequency")

sns.boxplot(x=df["sepal.length"], ax=axes[1], orient="h", width=0.3, palette="Set2")
axes[1].set_title("Box Plot of Sepal Length")
axes[1].set_xlabel("Sepal Length")

plt.show()
```

#### Write your insghts here

```python
# ( Histogram of Sepal Length) :

## Most sepal lengths are concentrated between approximately 5.0 to 6.5. 

## The histogram has a somewhat normal distribution shape with a slight central peak.
```

#### Write your insghts here

```python
# ( Box Plot of Sepal Length ) :

##The median sepal length is around 5.8 units.

##The box plot identifies a couple of shorter sepal length outliers compared to the majority of data points.
```

### Bivariate Visualization 
* Plot two Charts
* Provide each charts with at least two insights 

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="sepal.length", y="sepal.width", data=df, hue="variety", palette="Set2")
plt.title(" Sepal Length vs. Sepal Width")
plt.legend(title="Variety")

sns.pairplot(df, hue="variety", palette="Set2")
plt.suptitle("Numerical Columns", y=1.02)

plt.show()
```

#### Write your insghts here

```python
#(scatterplot):
1 #Different iris species exhibit distinct clusters.
2 #Sepal length and sepal width have a varying negative correlation across species.
```

#### Write your insghts here

```python
# (pairplot) :
1 # The data shows a clear capability to distinguish between different iris species based on their numerical features, making the data suitable for classification tasks.
2 # The data exhibits some relationships between numerical features, such as the negative relationship between sepal length and sepal width. In many instances, as sepal length increases, sepal width decreases, and vice versa.
```

### Multivariate Visualization 
* Plot two Charts
* Provide each charts with at least two insights 

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

colors = {'Setosa': 'r', 'Versicolor': 'g', 'Virginica': 'b'}

for species, color in colors.items():
    subset = df[df['variety'] == species]
    ax.scatter(subset['petal.length'], subset['petal.width'], subset['sepal.length'], c=color, label=species)

ax.set_xlabel('Petal Length')
ax.set_ylabel('Petal Width')
ax.set_zlabel('Sepal Length')
ax.set_title("3D Scatter Plot")
plt.legend()



plt.figure(figsize=(10, 6))
sns.violinplot(x="variety", y="sepal.length", data=df, palette="Set2")
plt.title("Violin Plot of Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length")

plt.show()
```

#### Write your insghts here

```python
# (3D scatter):
1# The 3D scatter plot distinguishes iris species based on petal length, petal width, and sepal length.
2# Setosa is characterized by shorter petal dimensions compared to other species.
```
